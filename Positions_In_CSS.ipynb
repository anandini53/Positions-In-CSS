2. **Task 2:**

 Problem Statement

 Explain the difference between Absolute and Relative positioning

**Answer:**

**Absolute Positioning:**
<ul>
  <li>In CSS, absolute positioning places an element at a fixed location relative to its closest positioned ancestor or the viewport.
  <li>It affects the position of other elements.
  <li> It does not retain its original space, other documents may occupy it.
  <li> It is used for exact positioning within a container or page.
  <li> It creates a new stacking context based on the ancestor's stacking order.
  <li> It is commomly use to precise positioning within a specific container or overlay effects.
</ul>

**Relative Positioning**
<ul>
  <li> Relative positioning allows an element to be moved from its original position in the flow of the page, but it still occupies space in the layout.
  <li> It doesn't affect the position of other elements.
  <li> It retains it original space in the document layout.
  <li> It is used to move an element slightly from its original position.
  <li> It can create a new stacking context if z-index is applied.
  <li> It does minor adjustments like nudging elements.

<ul>

5. **Task 5:**

 Problem Statement

 Explain the z-index, with a code example.


**Answer:**

When elements are positioned, they can overlap other elements.

The z-index property specifies the stack order of elements on a web page i.e. which element should be placed in front of, or behind the others.
An element can have a positive or negative stack order. Element with higher z-index value will appear in front of elements with a lower z-index element.


```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title> Z-index </title>

    <style>

        .parent {

            width: 200px;
            height: 200px;
            position: absolute;
            opacity: 0.8;

        }

        .box1 {

            background-color: rgb(253, 152, 169);
            top: 60px;
            left: 60px;
            z-index: 1;
        }

        .box2 {

            background-color: rgb(130, 226, 40);
            top: 110px;
            left: 110px;
            z-index: 3;
        }

        .box3 {

             background-color: rgb(40, 121, 226);
            top: 160px;
            left: 160px;
            z-index: 2;

        }

    </style>
</head>
<body>


    <div class="parent box1">
        Z-index: 1
    </div>

    <div class="parent box2">

        z-index:3

    </div>

     <div class="parent box3">
        Z-index: 2
    </div>


</body>
</html>
```
